```python
import pandas as pd
```

```python
dataset = pd.read_csv('covid_19_indonesia_time_series_all.csv')
dataset.head()
```

```python
dataset.drop(dataset.loc[dataset['Location']=='Indonesia'].index, inplace=True)
```

```python
dataset.head()
```

```python
dataset = dataset[['Date', 'Location', 'New Cases', 'New Deaths', 'New Recovered', 'New Active Cases', 'Total Cases', 'Total Deaths', 'Total Recovered', 'Longitude', 'Latitude', 'Growth Factor of New Cases', 'Growth Factor of New Deaths']]
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
dataset['Growth Factor of New Deaths'].fillna(value=dataset['Growth Factor of New Deaths'].mean(), inplace=True)
```

```python
dataset['Growth Factor of New Cases'].fillna(value=dataset['Growth Factor of New Cases'].mean(), inplace=True)
```

```python
dataset.head()
```

```python
dataset['Date']= pd.to_datetime(dataset['Date'])
```

```python
df_case=dataset.groupby(['Date'])['Total Cases'].agg('sum')
df_case=pd.DataFrame(df_case)
df_case.head()
```

```python
date= dataset['Date'].unique()
df_case['Date']=date
```

```python
X=df_case['Date']
Y=df_case['Total Cases']

import matplotlib.pyplot as plt
plt.figure(figsize=(6, 6))
plt.scatter(X,Y)
plt.xticks(rotation=45)
```

```python
from scipy.optimize import curve_fit
```

```python
df_jak=dataset[(dataset.Location =='DKI Jakarta')]
df_jak.head()
```

```python
import numpy as np
#country_data['GrowthFactor'] = growth_factor(country_data['Confirmed'])

# we will want x_data to be the number of days since first confirmed and the y_data to be the confirmed data. This will be the data we use to fit a logistic curve
x_data = range(len(df_jak['Date']))
y_data = df_jak['Total Cases']

def log_curve(x, k, x_0, ymax):
    return ymax / (1 + np.exp(-k*(x-x_0)))

# Fit the curve
popt, pcov = curve_fit(log_curve, x_data, y_data, bounds=([0,0,0],np.inf), maxfev=50000)
estimated_k, estimated_x_0, ymax= popt
```

```python
# Plot the fitted curve
k = estimated_k
x_0 = estimated_x_0
y_fitted = log_curve(range(0,500), k, x_0, ymax)
print(k, x_0, ymax)
#print(y_fitted)
y_data.tail()
```

```python
# Plot everything for illustration
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(range(0,500), y_fitted, '--', label='Predicted Case')
ax.plot(x_data, y_data, 'o', label='Real Case')
ax.set_xlabel('Days')
ax.set_ylabel('Total Cases')
ax.legend()
#ax.set_ylim(0,300000)
plt.show()
print('Day 0 is the earliest date in dataset')
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_data, y_fitted[:225])
```
